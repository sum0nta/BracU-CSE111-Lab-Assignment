```python
#Task2
class Shape:
    def __init__(self,shape,side1,side2):
        self.shape = shape
        self.side1 = side1
        self.side2 = side2
    def area(self):
        if self.shape == "Triangle":
            area = self.side1 * self.side2 * 0.5
        elif self.shape == "Square":
            area = self.side1 * self.side2
        elif self.shape == "Rhombus":
            area = self.side1 * self.side2 * 0.5
        elif self.shape == "Rectangle":
            area = self.side1 * self.side2
        else:
            area = "Shape Unknown"
        print(f"Area: {area}")
triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#Task3

class Calculator:
    def __init__(self):
        print("Calculator is ready!")
    def calculate(self,val1,val2,sign):
        self.val1, self.val2, self.sign = val1, val2, sign
    
        if self.sign == '+':
            self.result = self.val1 + self.val2
        elif self.sign == '-':
            self.result = self.val1 - self.val2
        elif self.sign == '*':
            self.result = self.val1 * self.val2
        elif self.sign == '/':
            self.result = self.val1 / self.val2
        return self.result
    def showCalculation(self):
        print(f"{self.val1} {self.sign} {self.val2} = {self.result}")

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#Task11

class TaxiLagbe:
    def __init__(self,numPlate,location):
        self.numPlate, self.location = numPlate, location
        self.passanger, self.total = [], 0
        
    def addPassenger(self,*passanger):
        for i in passanger:
            person,fare = i.split('_')
            if len(self.passanger) == 4:
                print("Taxi Full! No more passangers can be added.")
                break
            else:
                self.passanger.append(person)
                self.total += int(fare)
                print(f"Dear {person}! Welcome to TaxiLagbe.")

    def printDetails(self):
        print(f"Trip info for Taxi number: {self.numPlate}")
        print(f"The taxi can only cover {self.location} area.")
        print(f"Total passenger: {len(self.passanger)}")
        print("Passanger lists:")
        for i in self.passanger:
            if i != self.passanger[-1]:
                print(f"{i}, ",end = '')
            else:
                print(i)
        print(f"Total fare collected: {self.total}")
        
taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
  
```

```python
#Task13

class StudentDatabase:
    def __init__(self,name,idNum):
        self.name, self.id, self.cgpa = name, idNum, 0.0
        self.sem, self.grades = [], {}
    def calculateGPA(self,data,sem):
        self.courses,total_data = [], []
        total = 0
        if sem not in self.sem:
            for i in data:
                courseGrade = i.split(":")
                total_data.append(courseGrade[0])
                total_data.append(courseGrade[1].strip())
            courses, gpa = total_data[::2], total_data[1::2]
            gpa = [float(i) for i in gpa]
            for i in gpa:
                total += i
            self.cgpa = round(total/len(gpa),2)
            for i in courses:
                self.courses.append(i)
            d1 = {tuple(self.courses):self.cgpa}
            if sem not in self.grades:
                self.grades[sem] = d1  
        self.sem.append(sem)
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        for i in self.sem:
            print(f"Courses taken in {i}:")
            for key,values in self.grades[i].items():
                for j in key:
                    print(j)
                print(f"GPA: {values}")


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
#Task22

class Hotel:
    def __init__(self,name):
        self.name = name
        self.staff = []
        self.guest = []
    def addStuff(self,name,age,phone = '000'):
        self.staff.append({name:[age,phone]})
        print(f"Staff with ID {len(self.staff)} is added")
    def getStuffById(self,index):
        for name,data in self.staff[index-1].items():
            str1 = f"Staff ID: {index}\n"
            str1 += f"Name: {name}\n"
            str1 += f"Age: {data[0]}\n"
            str1 += f"Phone no. :{data[1]}"
        return str1

    def addGuest(self,name,age,phone):
        self.guest.append({name:[age,phone]})
        print(f"Guest with ID {len(self.guest)} is added")
    def getGuestById(self,index):
        for name,data in self.guest[index-1].items():
            str1 = f"Guest ID: {index}\n"
            str1 += f"Name: {name}\n"
            str1 += f"Age: {data[0]}\n"
            str1 += f"Phone no. :{data[1]}"
        return str1
    def allStaffs(self):
        print(f"All Staffs:")
        print(f"Number of Staff: {len(self.staff)}")
        str1 = ''
        for index in range(len(self.staff)):
            for name,data in self.staff[index].items():
                str1 += f"Staff ID: {index+1} "
                str1 += f"Name: {name} "
                str1 += f"Age: {data[0]} "
                str1 += f"Phone no. :{data[1]}"
                if index != (len(self.staff) - 1):
                    str1 += '\n'
        print(str1)

    def allGuest(self):
        print(f"All Guest:")
        print(f"Number of Guest: {len(self.guest)}")
        str1 = ''
        for index in range(len(self.guest)):
            for name,data in self.guest[index].items():
                str1 += f"Guest ID: {index+1} "
                str1 += f"Name: {name} "
                str1 += f"Age: {data[0]} "
                str1 += f"Phone no. :{data[1]}"
                if index != (len(self.guest) - 1):
                    str1 += '\n'
        print(str1)
h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest('Carol',35,'123')
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, '431')
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```
