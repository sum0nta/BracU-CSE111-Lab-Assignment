```python
#Task2

class Assassin:
    totalNum = 0
    def __init__(self,name,rate):
        self.name, self.rate = name,rate
        Assassin.totalNum += 1
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"Success rate: {self.rate}%")
        print(f"Total number of Assassin: {Assassin.totalNum}")
    @classmethod
    def failureRate(cls,name,rate):
        return cls(name,100-rate)
    @classmethod
    def failurePercentage(cls,name,rate):
        rate = 100 - int(rate[:-1])
        return cls(name,rate)
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#Task5

class Employee:
    def __init__(self,name,workingPeriod):
        self.name, self.workingPeriod = name, workingPeriod
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        return cls(name,2022-year)
    @staticmethod
    def experienceCheck(year,gender):
        if gender == 'male':
            if year < 3:
                return "He is not experienced"
            return "He is experienced"
        if year < 3:
            return "She is not experienced"
        return "She is experienced"
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task6

class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name, self.count = name,count
        Laptop.laptopCount += count
    @staticmethod
    def advantage():
        print("Laptops are portable")
    @classmethod
    def resetCount(cls):
        cls.laptopCount = 0
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#Task7

class Cat:
    Number_of_cats = 0
    def __init__(self,color,state):
        self.color, self.state = color, state
        Cat.Number_of_cats += 1
    @classmethod
    def no_parameter(cls):
        return cls("Black","sitting")
    @classmethod
    def first_parameter(cls,name):
        return cls(name,'sitting')
    @classmethod
    def second_parameter(cls,state):
        return cls("Grey",state)
    def printCat(self):
        print(f"{self.color} cat is {self.state}")
    def changeColor(self,color):
        self.color = color

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task8
from math import pi
class Cylinder:
    radius, height = 5,18
    def __init__(self,radius,height):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        self.radius, self.height = radius, height
        if Cylinder.radius != self.radius or Cylinder.height != self.height:
            print(f"Updated: radius={self.radius} and height={self.height}.")
        Cylinder.radius, Cylinder.height = self.radius, self.height
    @classmethod
    def changeFormat(cls,data):
        radius,height = data.split("-")
        return cls(float(radius),float(height))
    @classmethod
    def swap(cls,height,radius):
        #cls.height, cls.radius = height, radius
        return cls(radius,height)
    @staticmethod
    def area(radius,height):
        area = (2 * pi* (radius**2)) + (2* pi* radius * height)
        print(f"Area: {area}")
    @staticmethod
    def volume(radius,height):
        volume = pi * (radius ** 2) * height
        print(f"Volume : {volume}")

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task10

class SultansDine:
    branchNum = 0
    totalSell = 0
    all_info = []
    def __init__(self,location):
        self.location = location
        self.sellNumber = 0
        self.sellTotal = 0
        SultansDine.all_info.append(self)
        SultansDine.branchNum += 1
    def sellQuantity(self,number):
        self.sellNumber += number
    def branchInformation(self):
        if self.sellNumber < 10:
            branch_sell = self.sellNumber * 300
        elif self.sellNumber < 20:
            branch_sell = self.sellNumber * 350
        else:
            branch_sell = self.sellNumber * 400
        self.sellTotal += branch_sell
        SultansDine.totalSell += branch_sell
        print(f"Branch Name: {self.location}")
        print(f"Branch Sell: {branch_sell} Taka")
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.branchNum}")
        print(f"Total Sell: {cls.totalSell} Taka")
        for i in cls.all_info:
            print(f"Branch Name: {i.location}, Branch Sell: {i.sellTotal} Taka")
            percent = ((i.sellTotal) / cls.totalSell) * 100
            print(f"Branch consists of total sell's: {round(percent,2)} %") 
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
